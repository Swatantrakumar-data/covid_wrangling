```python
data = data.loc[:,['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

```python
 data.head()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=['continent'],inplace=True)
```

```python
data.fillna(value=0,inplace=True)
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data['month'] = pd.to_datetime(data['date']).dt.month
```

```python
data
```

```python
data.groupby('continent').max()
```

```python
df_groupby = data.reset_index()
```

```python
df_groupby.head(10)
```

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
```

```python
sns.distplot(df_groupby['gdp_per_capita'],kde=False,bins=50)
```

```python
sns.scatterplot(x = 'total_cases',y = 'gdp_per_capita',data=df_groupby)
```

```python
sns.pairplot(df_groupby)
```
